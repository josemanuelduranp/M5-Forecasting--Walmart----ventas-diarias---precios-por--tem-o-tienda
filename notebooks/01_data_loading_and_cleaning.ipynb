```python
#Selección de subset: Se trabajará con un subset controlado...
#Estrategia:
#2-3 categorías (dept_id)
#3 estados-> luego se mapeará por países

selected_depts = ['FOODS_3', 'HOBBIES_1']
selected_states = ['CA', 'TX', 'WI']

sales_subset = sales_val[
    (sales_val['dept_id'].isin(selected_depts)) &
    (sales_val['state_id'].isin(selected_states))
]
```
